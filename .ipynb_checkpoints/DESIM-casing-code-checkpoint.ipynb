```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
```

# Forecasting France's GDP

## Loading the data

```python
# Load the Excel file
gdp_file_path = 'API_NY.GDP.MKTP.CD_DS2_en_excel_v2_93.xlsx'
gdp_data = pd.read_excel(gdp_file_path, sheet_name="Data", skiprows=3, engine='openpyxl')
france_gdp = gdp_data[gdp_data['Country'] == 'France']

# USED FOR LATER
france_gdp_2022 = france_gdp['2022']

france_gdp.head()
```

```python
france_gdp_years = france_gdp.loc[:, '1960':'2022'].transpose()
france_gdp_years.columns = ['GDP']
france_gdp_years.index = france_gdp_years.index.map(int) 

dates = pd.date_range(start='1960', periods=len(france_gdp_years), freq='AS')  # 'AS' for annual start frequency
france_gdp_years.index = dates
```

```python
plt.figure(figsize=(10, 6))
plt.plot(france_gdp_years.index, france_gdp_years['GDP'], marker='o', color='blue')
plt.title('France GDP Over Time (1960-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (current US$)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
```

## Applying ARIMA model to forecast France's GDP

### Checking stationarity

```python
fractional = france_gdp_years.pct_change()
fractional = fractional.iloc[1:]

fractional.plot(figsize=(10,6), marker='o')
plt.ylabel('Fractional change in France GDP')
plt.grid()
plt.show()
```

```python
# Perform augmented Dickey-Fuller test to check if returns are stationary
pvalue = adfuller(fractional)[1]
if pvalue < 0.05:
    print('Returns are stationary --> d = 1')
else:
    print('Returns are non-stationary --> d > 1')
```

### Estimating the number of Autoregressive terms and Moving Average terms

```python
# Check the ACF plot to estimate parameter q
plot_acf(fractional)
plt.show()
```

```python
# Check the PACF plot to estimate parameter p
plot_pacf(fractional)
plt.show()
```

Since the only significant lags are at 1 on both the ACF and PACF, we can conclude that that **p** and **q** could both equal 1.

### Error analysis using train/test sets

```python
france_gdp = france_gdp_years['GDP']
split_point = int(len(france_gdp) * 0.8) 
train, test = france_gdp[0:split_point], france_gdp[split_point:]

print(f"train size: {len(train)}. test size: {len(test)}")
```

```python
# Fit ARIMA model
model = ARIMA(train, order=(1,1,1))  # (1, 1, 1) found through analysis above
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test))
forecast_series = pd.Series(forecast, index=test.index)
```

```python
mae = mean_absolute_error(test, forecast_series)
mse = mean_squared_error(test, forecast_series)
rmse = np.sqrt(mse)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
residuals = test - forecast_series

# Plot residuals
plt.figure(figsize=(10,6))
plt.plot(residuals)
plt.title('Residuals from ARIMA Model', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Error', fontsize=12)
plt.axhline(0, color='red', linestyle='--')
plt.show()

# Plot histogram to check the distribution of residuals
plt.figure(figsize=(10,6))
plt.hist(residuals, bins=20)
plt.title('Distribution of Residuals', fontsize=14)
plt.xlabel('Error', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

# Ljung-Box test on residuals to check for any autocorrelation
lb_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
print(lb_test)
```

# Creating a Gravitational Model of Trade between a subset of EU countries

## Data processesing

```python
eu_members_path = 'list_of_EU_countries.xlsx'
eu_members = pd.read_excel(eu_members_path)
```

```python
capitals_locations_path = 'country-capital-lat-long-population.csv'
capitals_locations = pd.read_csv(capitals_locations_path)
eu_locs = capitals_locations[capitals_locations['Country'].isin(eu_members['Countries'])]
eu_locs = eu_locs[['Country', 'Latitude', 'Longitude']]
```

```python
from geopy.distance import geodesic

def dist_to_france(lat, long):
    # Coordinates for Paris, France
    paris_coords = (48.8534, 2.3488)
    
    other_coords = (lat, long)
    
    # Calculate and return the distance in miles
    return geodesic(paris_coords, other_coords).miles
```

```python
eu_locs['Distance to France (Miles)'] = eu_locs.apply(lambda x: dist_to_france(x['Latitude'], x['Longitude']), axis=1)
eu_dist = eu_locs[['Country', 'Distance to France (Miles)']].copy()
eu_dist['log Distance to France (Miles)'] = np.log(eu_dist['Distance to France (Miles)'] + 1)
eu_dist.head()
```

```python
eu_gdp_years = gdp_data[gdp_data['Country'].isin(eu_members['Countries'])]
eu_gdp = eu_gdp_years[['Country', '2022']].copy()
eu_gdp['log 2022'] = np.log(eu_gdp['2022'] + 1)
eu_gdp.head()
```

```python
exports_path = 'exports_from_country_to_france.xlsx'
imports_path = 'imports_from_france_to_country.xlsx'

exports_years = pd.read_excel(exports_path, skiprows=6, engine='openpyxl')
imports_years = pd.read_excel(imports_path, skiprows=6, engine='openpyxl')

exports_years = exports[exports['Country'].isin(eu_members['Countries'])]
imports_years = imports[imports['Country'].isin(eu_members['Countries'])] 

imports = imports_years[['Country', '2022']].copy()
exports = exports_years[['Country', '2022']].copy()
```

```python
# Calculate total trade volume (sum of exports and imports for each country)
total_trade = exports.merge(imports, on='Country', suffixes=('_export', '_import'))
total_trade['Trade (USD millions)'] = total_trade['2022_export'] + total_trade['2022_import']
total_trade['log Trade (USD millions)'] = np.log(total_trade['Trade (USD millions)'] + 1)
total_trade.head()
```

```python
data_tmp = pd.merge(total_trade, eu_gdp, on='Country')
data_tmp2 = pd.merge(data_tmp, eu_dist, on='Country')
data = data_tmp2[['Country', 'log Trade (USD millions)', 'log 2022', 'log Distance to France (Miles)']].copy()
data.columns = ['country', 'log trade', 'log other gdp', 'log distance']
data['log france gdp'] = np.log(float(france_gdp_2022) + 1)
data.head()
```

## Model training

```python
X = data[['log other gdp', 'log france gdp', 'log distance']]
y = data['log trade']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
```

## Error analysis

```python
print(model.summary())
```

```python
data['residuals'] = model.resid

# Plotting residuals
plt.scatter(data['log distance'], data['residuals'])
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Log Distance')
plt.ylabel('Residuals')
plt.show()
```

```python
from geographiclib.geodesic import Geodesic

def calc_halfway_coords(lat1, long1, lat2, long2, howMuchBetween):
    geod = Geodesic.WGS84

    g = geod.Inverse(lat1, long1, lat2, long2)
    l = geod.Line(g['lat1'], g['lon1'], g['azi1'])
    
    # find midpoint
    midpoint = l.Position(howMuchBetween * g['s12'])

    return midpoint

# closerToLat and closerToLong are the coordinates corresponding to the destination that France moves toward (in this case London)
# howMuchCloser is from [0, 1] indicates by how much we move toward the location (in this case 0.5)
# lat and long are the coords of the other country 
def dist_to_relocated_france(closerToLat, closerToLong, howMuchCloser, lat, long):
    relocated_coords = calc_halfway_coords(48.8534, 2.3488, closerToLat, closerToLong, howMuchCloser)

    other_coords = (lat, long)
    
    # Calculate and return the distance in miles
    return geodesic(paris_coords, other_coords).miles

print(dist_to_relocated_france(51.509865, -0.118092, 51.509865, -0.118092, 0.5))
```
